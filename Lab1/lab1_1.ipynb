```python
review=positive_reviews[107]
print(review)
```

```python
nltk.download('stopwords')
```

```python
import re
import string 
from nltk.corpus import stopwords 
from nltk.stem import PorterStemmer 
from nltk.tokenize import TweetTokenizer
```

```python
print('\033[92m' + review)
print('\033[94m')
review2 = re.sub(r'https?:\/\/.*[\r\n]*', '', review)
review2 = re.sub(r'#', '', review)
print(review2)
```

```python
print()
print('\033[92m' + review2)
print('\033[94m')
tokenizer = TweetTokenizer(preserve_case=False)
review_tokens = tokenizer.tokenize(review2)
print()
print('Tokenized string:')
print(review_tokens)
```

```python
stopwords_english = stopwords.words('english')
print('Stop words\n')
print(stopwords_english)
print('\nPunctuation\n')
print(string.punctuation)
```

```python
print()
print('\033[92m')
print(review_tokens)
print('\033[94m')
reviews_clean = []
for word in review_tokens: 
  if (word not in stopwords_english and 
    word not in string.punctuation): 
    reviews_clean.append(word)
print('removed stop words and punctuation:')
print(reviews_clean)
```

```python
print()
print('\033[92m')
print(reviews_clean)
print('\033[94m')
stemmer = PorterStemmer()
reviews_stem = []
for word in reviews_clean:
  stem_word = stemmer.stem(word)
  reviews_stem.append(stem_word)
print('stemmed words:')
print(reviews_stem)
```
